#### Predict canada's per capita income in year 2020. Using csv file build a regression model and predict the per capita income for canadian citins in year 2020.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
db1 = pd.read_csv('canada.csv')
db1
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Annual Income')
plt.scatter(db1.year,db1.annual_income, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(db1[['year']],db1.annual_income)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict(np.array([[2020]]))
```
